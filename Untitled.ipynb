```python
import numpy as np
import pandas as pd
import unidecode
import csv
```

```python
video_df = pd.read_csv('bkr_channel_data.csv')
video_df
```

```python
video_df['title'] = video_df['title'].apply(lambda x: unidecode.unidecode(x))
video_df
```

```python
video_df['description'] = video_df['description'].apply(lambda x: unidecode.unidecode(x))
video_df
```

```python
video_df.iloc[27,3]
```

```python
video_df.to_csv('bkr_channel_data_clean.csv', quoting=csv.QUOTE_NONNUMERIC,escapechar="\\",doublequote=False,index=False)
```
